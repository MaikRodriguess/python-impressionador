# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
print('Gere 50 números aleatórios de 1000 a 3000:\n{} '.format(vendas))
print('Retorna valores dentro de um intervalo de 1 a 50: \n', meses)
```

```python
plt.plot(meses,vendas)
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses,vendas,'-X') # Gráfico em linhas com marcação em X 
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize= (15, 3))
# subplot(linha, coluna, índice)
plt.subplot(1, 3, 1) 
plt.plot(meses, vendas, color = 'blue')
plt.subplot(1, 3, 2)
plt.scatter(meses, vendas, color = 'green')
plt.subplot(1, 3, 3)
plt.bar(meses, vendas, color = 'red')
plt.show()
```
